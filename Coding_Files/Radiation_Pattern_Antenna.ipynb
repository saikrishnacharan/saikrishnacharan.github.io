```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
T = 180
```

```python
def vary_par(ratio,N):
    T=180
    theta = np.linspace(-90,90,T)
    pd=2*np.pi*ratio*(np.sin(np.deg2rad(theta)))
    AF=np.zeros(T)+1j*np.zeros(T)
    for j in range(T):
        for i in range(N):
            AF[j]=AF[j]+np.exp(1j*i*pd[j]);
    return AF
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter
import numpy as np
from celluloid import Camera
# from matplotlib.animation.PillowWriter
fig = plt.figure(figsize=(8, 5))
# fig=plt.figure()
plt.xlim(-90,90)
plt.ylim(-40,1)
camera = Camera(fig)
# val=np.linspace(0.5,5,20)
val=np.arange(0.25,5,0.25)
for i in val:
    T=180
    AF_th=vary_par(i,2)
    Angles = np.linspace(-90,90,T)
    f = plt.plot(Angles,10*np.log10((abs(AF_th*AF_th))/4))
    plt.ylabel("AF(theta)")
    plt.xlabel("theta")
    plt.title("Radiation Patter by tweaking d/Î» with 2 antennas")
    plt.legend(f, [f'd/Î» = {i}'])
    camera.snap()
animation1 = camera.animate(interval = 1000, repeat = True,repeat_delay = 1000,blit=True)
f = r"C:\charan/Ratio_Tweak.gif" 
writergif = PillowWriter(fps=1) 
animation1.save(f, writer=writergif)
```
